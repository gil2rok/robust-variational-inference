```python
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt
```

```python
def gaussian_mixture(k=2, sigma_sqr=4, n=100000):
    '''
    Calculate a gaussian mixture model.
    Inputs:
        k - number of Gaussian disstributions that we will draw from
        sigma_sqr -  variance used in selecting the mean of our 
                    different models
        n - number of samples
    
    Returns:
        vector of n samples drawn from the guassian mixture model specified
        from the input parameters
    '''
    
    # Draw the k means of our Gaussian distributions from a Gaussian distribution
    std_dev = np.sqrt(sigma_sqr)
    mew_vec = default_rng().normal(0, std_dev, k) # draw k samples from Gaussian distribution
    mew = np.tile(mew_vec, n).reshape((n, mew_vec.size)) # repeat to enable code vectorization
    
    # Draw n samples from our Gaussian mixture model
    c_matrix = np.zeros((n,k))
    c_index = np.random.randint(0,k,n) # uniformly pick a model
    c_matrix[np.arange(n), c_index, ]=1 # store model info in one-hot encoding matrix
    c_mean = np.multiply(c_matrix, mew).sum(axis=1) # Find mean for given models
    x = np.random.default_rng().normal(c_mean, 1) # Draw from a Gaussian model w/ relevant mean and std dev of 1
    
    #print("Mew_Vec:\n", mew_vec, "\n")
    #print("C_Index:\n", c_index, "\n")
    #print("C_Matrix:\n", c_matrix, "\n")
    #print("C_Mean:\n", c_mean, "\n")
    
    return x, mew_vec, c_matrix, c_mean
```

```python
# Pick hyper-parameters
k = 2; sigma_sqr = 4

# Run model and get 'x' data points, 'mew' means, and 'c_matrix' vector encodings of mixture components
x, mu, c_matrix, mean_assignements = gaussian_mixture(k=k, sigma_sqr=sigma_sqr)

plt.hist(x, bins='auto')
plt.show()
```

```python
def ELBO(m, s_sqr, phi, sigma_sqr, x):
    n = np.size(x) # Number of data points
    K = np.size(m) # Number of different clusters
    
    assert(K > 0) # Need to calculate np.log(k)
    assert(sigma_sqr > 0) # Need to calculate np.log(sigma_sqr)
    
    #print("M: ", m, "\nS_sqr: ", s_sqr, "\nPhi: ", phi, "\n\n")
    
    # Term 1
    term1 = (K*0.5)*(1-np.log(2*np.pi*sigma_sqr)) - n*np.log(K * np.power(2*np.pi, 0.5))
    
    # Term 2
    term2 = 0
    for k in range(K):
        term2 += np.log(2*np.pi*s_sqr[k]) - ((s_sqr[k] + m[k]**2)/sigma_sqr)
    term2 *= 0.5
    
    # Term 3
    term3 = 0
    for i in range(n):
        for k in range(K):
            term3 += phi[i,k]*((x[i]**2) - (2*x[i]*m[k]) + s_sqr[k] + (m[k]**2) + (2*np.log(phi[i,k])))
    term3 *= -0.5
    
    #print("Term 1: ", term1, " Term 2: ", term2, " Term 3: ", term3, "\n\n")
    return term1 + term2 + term3
```

```python
def init_params(x, K, seed=12345):
    '''
    Randomly initialize the parameters m, s_sqr, and phi for
    the CAVI algorithm by "drawing from a factorized Gaussian
    calibrated to the empirical mean and variance of the dataset"
    
    Inputs:
        x - data in an n-length vector
        K - number of cluster assignements
    
    Returns:
        m - a K-length vector of initialized means, with each
            element drawn as m_k ~ N(mean(data, std_dev(data)))
        s_sqr - a K-length vector of initialized variance, with
                each element drawn as 
                (s_sqr)_k ~ N(std_dev(data), std_dev(std_dev(data)))
        phi - a Nxk matrix, with each k-length row vector being
                an indicator vector for cluster assignements.
                Each phi_k initialized as phi_k ~ U(1/k)
    '''
    n = np.size(x)
    rng = default_rng(seed) # pass seed value to default_rng
    
    m = rng.normal(loc=np.mean(x), scale=np.std(x), size=(K))
    #s_sqr = rng.normal(loc=np.std(x), scale=np.std(np.std(x)), size=(K))
    #s_sqr = rng.normal(loc=np.std(x), scale=1, size=(K,1))
    s_sqr = np.ones((K))
    phi = np.full((n,K), 1/K) # Each row is a uniform, k-length indicator-vector that sums to 1
    return m, s_sqr, phi

m, s_sqr, phi = init_params(x, k)
```

```python
def update_params(m, s_sqr, phi, x):
    n = np.size(x) # Number of data points
    K = np.size(m) # Number of different clusters
    
    # Update phi
    for i in range(n):
        normalizing_const = 0
        for k in range(K):
            phi[i,k] = np.exp(m[k]*x[i] - ((s_sqr[k]+m[k]**2) * 0.5))
            normalizing_const += phi[i,k]
       
        for k in range(K):
            phi[i,k] /= normalizing_const

    # Update m and s_sqr
    for k in range(K):
        t1 = 0; t2 = 0
        for i in range(n):
            t1 += phi[i,k] * x[i]
            t2 += phi[i,k]

        m[k] = t1 / (t2 + 1/sigma_sqr)
        s_sqr[k] = 1 / (t2 + 1/sigma_sqr)
        
    return m, s_sqr, phi
```

```python
def CAVI(x, K, sigma_sqr, tol=1e-4, max_iter=20):
    '''
    Calculating the CAVI for a gaussian mixture model
    Inputs:
        x - data in n-length vector
        K - number of cluster assignements
        sigma_sqr - prior variance of component means
        
    Returns:
        Approximations of mean, variance, and cluster assignements of gaussian mixture model
        m - a K-length vector of mean approximations
        s_sqr - a K-length vector of variance approximations
        phi - a Nxk matrix, with each k-length row vector being
                an indicator vector for cluster assignements.
    '''
    
    ELBO_list = []
    n = x.size # number of data points
        
    # Init m, s_sqr, and phi and calculate ELBO
    m, s_sqr, phi = init_params(x,K)
    prev_ELBO = ELBO(m, s_sqr, phi, sigma_sqr, x)
    print("Cur ELBO: ", prev_ELBO, "\n")
   
    converged = False
    iter_num = 0
    
    while converged == False and iter_num < max_iter:
        
        #Update parameters
        m, s_sqr, phi = update_params(m, s_sqr, phi, x)
    
        # Compute new ELBO
        cur_ELBO = ELBO(m, s_sqr, phi, sigma_sqr, x)

        # Check if converged
        diff = np.abs(cur_ELBO - prev_ELBO)
        ELBO_list.append(prev_ELBO)
        
        if (diff < tol):
            converged = True
        else:
            prev_ELBO = cur_ELBO
        
        iter_num += 1
       
        print("Iter Num: ", iter_num)
        print("Cur ELBO: ", cur_ELBO)
        print("Diff: ", diff, "\n")

    return m, s_sqr, phi, prev_ELBO, ELBO_list
```

```python
m, s_sqr, phi, prev_ELBO,ELBO_list = CAVI(x, k, sigma_sqr)
```

```python
def test(mu, c_matrix, m, phi):
    n = np.shape(phi)[0]
    K = np.size(mu)
    
    # Calc Mean Assignements
    mu_repeat = np.tile(mu, (n,1))
    mean_assignements = mu[np.argmax(c_matrix, axis=1)]
    
    # Calc Approx Mean Assignements
    m_repeat = np.tile(m, (n,1))
    approx_mean_assignements = m[np.argmax(phi, axis=1)]
    
    # Create dict that matches mean values to cluster number
    mu_sorted = np.sort(mu)
    mu_index = np.arange(np.size(mu))
    cluster_dict = dict(zip(mu_sorted, mu_index))
    
    #Create approx dict that matches approx mean values to cluster number
    m_sorted = np.sort(m)
    m_index = np.arange(np.size(m))
    approx_cluster_dict = dict(zip(m_sorted, m_index))
    
    # Plug mean and approx mean assignements into respective dictionaries to find cluster assignements
    cluster_assignements = np.zeros((n))
    approx_cluster_assignements = np.zeros((n))
    
    for i in range(n):
        cluster_assignements[i] = cluster_dict[mean_assignements[i]]
        approx_cluster_assignements[i] = approx_cluster_dict[approx_mean_assignements[i]]
    
    # Calc num of correct appox cluster assignements
    num_correct = np.count_nonzero(cluster_assignements == approx_cluster_assignements)
    percent_correct = num_correct/n
    
    # Num percentage points better than random guessing
    random_improvement = np.abs(percent_correct - 1/K) * 100
    
    
    #print("Mean Assignements:\n", mean_assignements, "\n")
    #print("Approx Mean Assignements:\n", approx_mean_assignements, "\n")
    #print("\nCluster Assignements:", cluster_assignements, sep="\n")
    #print("", "Approx Cluster Assignements,", approx_cluster_assignements, sep="\n")
    print(f"{percent_correct}% of cluster assignements predicted correctly\n")
    print(f"{random_improvement:.2f} percentage points better than randomly guessing\n")
    
    return percent_correct, random_improvement
```

```python
#perc_correct = test(mu, c_matrix[:10,:], m, phi[:10,:])
percent_correct, random_improvement = test(mu, c_matrix, m, phi)
```
